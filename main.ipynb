# Data Feature Engineering

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,LabelEncoder
import pickle
```

```python
df=pd.read_csv("Bank Customer Churn Prediction.csv")
df.head()
```

```python
df=df.drop('customer_id',axis=1)
df
```

```python
#For binary categorical data
gender_lbl_enc=LabelEncoder()

df['gender']=gender_lbl_enc.fit_transform(df['gender'])
df
```

```python
#For non binary categorical data we use OnehotEncoding
from sklearn.preprocessing import  OneHotEncoder

country_one_hot_encoder=OneHotEncoder()
country_ohe=country_one_hot_encoder.fit_transform(df[['country']])
country_ohe.toarray()
```

```python
country_encoded_df=pd.DataFrame(country_ohe.toarray(),columns=country_one_hot_encoder.get_feature_names_out(['country']))
country_encoded_df
```

```python
df=pd.concat([df.drop('country',axis=1),country_encoded_df],axis=1)
df
```

```python
#Save encoders in pickle file
with open('label_encoder.pkl','wb') as file:
    pickle.dump(gender_lbl_enc,file)

with open('ohe.pkl','wb') as file:
    pickle.dump(country_one_hot_encoder,file)
```

```python
x=df.drop('churn',axis=1)
y=df['churn']

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)

#Scaling the features
scaler=StandardScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.fit_transform(x_test)

```

```python
with open('scale.pkl','wb+') as file:
    pickle.dump(scaler,file)
```

# ANN Implementation

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.callbacks import EarlyStopping,TensorBoard
import datetime

log_dir="logfile/"+datetime.datetime.now().strftime('%Y%m%d-%H%M%S')
tensorboard=TensorBoard(log_dir=log_dir,histogram_freq=1)

early_stopping=EarlyStopping(monitor='val_loss',patience=7,restore_best_weights=True)
# IF MY LOSS VALUES REMAINS CONSTANT AFTER CONTINOUS BACKWARD AND FORWARD OPERATION THEN WE EARLYSTOP THE TRAINING
```

```python
#Building ANN Model
ann_model=Sequential([
    Dense(64,activation='relu',input_shape=(x_train.shape[1],)), #HL1 where input shape is in form of a tuple
    Dense(32,activation='relu'),#HL2
    Dense(1,activation='sigmoid')# Output Layer
])
```

```python
ann_model.summary()
```

```python
#Model compilation
import tensorflow
opt=tensorflow.keras.optimizers.Adam(learning_rate=0.01)
loss=tensorflow.keras.losses.BinaryCrossentropy()

ann_model.compile(optimizer=opt,loss=loss,metrics=['accuracy'])
```

# Training the model

```python
data_av=ann_model.fit(
    x_train,y_train,validation_data=(x_test,y_test),
    callbacks=[tensorboard,early_stopping],epochs=100
)
```

```python
ann_model.save('my_model.h5')
```
